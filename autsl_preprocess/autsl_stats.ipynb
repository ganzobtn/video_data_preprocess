```python
import os
```

```python
data_path = '/home/ganzorig/Datas/chalearn_processed_full/color/'
```

```python
import random
import shutil

dest_path = '/home/ganzorig/Datas/chalearn_processed_sample'
for clas in os.listdir(os.path.join(data_path,'train')):
    videos = os.listdir(os.path.join(data_path,'train',clas))
    sample_video = random.choice(videos)
    shutil.copy(os.path.join(data_path,'train',clas,sample_video), os.path.join(dest_path,clas+'.mp4'))
    
```

```python
data = {}
for split in ['train','test','val']:
    data[split] = {}
    for clas in sorted(os.listdir(os.path.join(data_path,split))):
        print(clas)
        data[split][clas] = {i: 0 for i in range(50)}
        videos = os.listdir(os.path.join(data_path,split,clas))
        #video_count = {i: 0 for i in range(50)}
        for video in videos:
            person_number = int(video.split('_')[0][6:])
            #video_count[person_number] += 1
            data[split][clas][person_number]+=1 
            #print(video)
            #print(person_number)
```

```python
import matplotlib.pyplot as plt

for split in ['train','test','val']:
    for clas in data[split]:
        #plt.plot(data[split][clas])
        a = []
        for i in data[split][clas]:
            if data[split][clas][i]==0:
                a.append(i)
        for i in a:
            data[split][clas].pop(i)
        #lists = sorted(data[split][clas].items()) # sorted by key, return a list of tuples

        #x, y = zip(*lists) # unpack a list of pairs into two tuples

        #plt.plot(x, y)
        #plt.show()
```

```python
count = {i: 0 for i in range(50)}

for i in data['train']:
    #print(i)
    #print(data['train'][i])
    for k in data['train'][i]:
        count[k]+=1

a = []
for i in count:
    if count[i]==0:
        a.append(i)
for i in a:
    count.pop(i)

fig = plt.figure(figsize = (10, 5))
x = list(count.keys())
y = list(count.values())

# creating the bar plot
plt.bar(range(len(count)),y , color ='maroon',
         width = 0.4)


plt.xlabel("signer")
plt.ylabel("No. of samples")
plt.title('Number of samples per signer in training set ')
plt.show()
```
