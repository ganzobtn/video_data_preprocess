```python
import pickle
import numpy as np
import os
```

```python
keypoint_path = '../data/wlasl_2000/keypoints_hrnet_dark_coco_wholebody.pkl'
hands_path = 'output_files'
```

```python
with open(keypoint_path, 'rb') as f:
    keypoints = pickle.load(f)
```

```python
hand_keypoints_range = list(range(91, 133))
```

```python
for key in keypoints.keys():
    print(keypoints[key].shape)
    hand_array = np.load(os.path.join(hands_path,key+'.npy'))*256
    print(hand_array.shape)
    
    keypoints[key][:,91:133,:] = hand_array
```

```python
with open('keypoints_hrnet_dark_coco_wholebody_mediapipe_hand.pkl', 'wb') as handle:
    pickle.dump(keypoints, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
